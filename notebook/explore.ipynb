```python
def prepare_serie(data_df):
    data_df = data_df.copy()
    data_df["Date"] = pd.to_datetime(data_df["Date"])
    data_df = data_df.sort_values("Date")
    data_df = data_df.set_index("Date")
    data_df = data_df.asfreq("B")
    data_df["Price"] = data_df["Price"].ffill()
    return data_df

X_data_process_df = prepare_serie(X_data_process_df)
Y_data_process_df = prepare_serie(Y_data_process_df)
Z_data_process_df = prepare_serie(Z_data_process_df)
```

### Sección 1. Modelado con log-retornos

```python
def calculate_log_return(data_df):
    data_df = data_df.copy()
    data_df["log_return"] = np.log(data_df["Price"] / data_df["Price"].shift(1))
    return data_df.dropna()

X_data_process_df = calculate_log_return(X_data_process_df)
Y_data_process_df = calculate_log_return(Y_data_process_df)
Z_data_process_df = calculate_log_return(Z_data_process_df)

print(X_data_process_df.shape)
print(Y_data_process_df.shape)
print(Z_data_process_df.shape)
```

#### Modelos preliminares

Se pronosticará 1 año para evaluar el desempeño del modelo; al tratar únicamente días hábiles, se proyectarán 252 observaciones. A los precios históricos se les aplicó transformación de log-retorno con el fin de estabilizar la varianza y garantizar estacionariedad, validada mediante la prueba ADF (Augmented Dickey-Fuller) anteriormente realizara con las materias primas y el cálculo de p-value.

Para ser eficientes en tiempo, vamos a realizar dos momentos en la evaluación de los modelos:
1. Se realizara una evaluación con el modelo ARIMA con múltiples parámetros de acuerdo a la métrica AIC que es recomendada para más rápida validación de mejores modelos e hiperparámetros, frente a otras métricas.
2. Con los mejores 3 hiperparámetros, entrenaremos el modelo ARIMA (comúnmente mejor que ETS para estos casos) y pasaremos a pronosticar con las 252 observaciones.
   
    Seleccionaremos el mejor y continuaremos en la sección del pronostico de los 36 meses.

_No trataremos modelos como SARIMAX debido a que los log-retornos son estacionarios (ADF p-value < 0.05) y no hay una demostración de estacionalidad significativa._

Los modelos se compararán utilizando:
1. MAE: Error absoluto promedio.
2. RMSE: Penaliza con mayor peso los errores grandes.

#### Funciones

```python
def select_by_aic(serie_log, p_values, q_values):

    serie_log = serie_log.dropna().reset_index(drop=True)
    results = []

    for p in p_values:
        for q in q_values:
            try:
                model = ARIMA(serie_log, order=(p,0,q))
                result = model.fit(method="innovations_mle")
                results.append((p,q,result.aic))
            except:
                continue

    data_df = pd.DataFrame(results, columns=["p","q","AIC"])
    return data_df.sort_values("AIC").reset_index(drop=True)
```

```python
def rolling_cv_top_models(
    serie_log, model_list, initial_train_size=0.6, forecast_horizon=21, step=21):
    serie_log = serie_log.dropna().reset_index(drop=True)
    n = len(serie_log)
    train_size = int(n * initial_train_size)
    results = []

    for (p,q) in model_list:
        maes = []
        rmses = []
        start = train_size
        print(p,q)

        while start + forecast_horizon <= n:
            train = serie_log.iloc[:start]
            test = serie_log.iloc[start:start+forecast_horizon]
            try:
                model = ARIMA(train, order=(p,0,q))
                result = model.fit(method="innovations_mle")
                forecast = result.forecast(steps=forecast_horizon)

                mae = mean_absolute_error(test, forecast)
                rmse = np.sqrt(mean_squared_error(test, forecast))
                maes.append(mae)
                rmses.append(rmse)
            except:
                break
            start += step
        if maes:
            results.append({
                "p": p,
                "q": q,
                "MAE": np.mean(maes),
                "RMSE": np.mean(rmses)
            })

    return pd.DataFrame(results).sort_values("RMSE").reset_index(drop=True)
```

##### Reconstrucción de los precios desde log-retornos

Tal como calculamos los log-retornos a partir de los precios, ahora realizaremos lo inverso para tener de nuevo los precios.
Fórmula:
`Pt​=Pt−1​e^(rt)​`

```python
def reconstruct_prices(last_price, forecast_log):
    prices = [last_price]
    for r in forecast_log:
        next_price = prices[-1] * np.exp(r)
        prices.append(next_price)
    return prices[1:]
# X
forecast_X_prices = reconstruct_prices(X_data_process_df["Price"].values[-1], forecast_X_log)
# Y
last_real_price_Y = Y_data_process_df["Price"].iloc[-1]
gap_prices_Y = reconstruct_prices(
    last_real_price_Y,
    forecast_Y_gap
)
last_aligned_price_Y = gap_prices_Y[-1]
forecast_Y_prices = reconstruct_prices(last_aligned_price_Y, forecast_Y_log)
# Z
last_real_price_Z = Z_data_process_df["Price"].iloc[-1]
gap_prices_Z = reconstruct_prices(
    last_real_price_Z,
    forecast_Z_gap
)
last_aligned_price_Z = gap_prices_Z[-1]
forecast_Z_prices = reconstruct_prices(last_aligned_price_Z, forecast_Z_log)
```

##### Aplicación de la segunda parte (Análisis de escenarios)

Para esto aplicaremos la fórmula de Monte Carlo

```python
def monte_carlo_arima_pro(
    model, last_price, steps, simulations=1000):
    simulated_price_paths = []

    for i in range(simulations):
        simulated_returns = model.simulate(
            nsimulations=steps
        )
        prices = [last_price]

        for r in simulated_returns:
            prices.append(prices[-1] * np.exp(r))

        simulated_price_paths.append(prices[1:])

    return np.array(simulated_price_paths)

monte_carlo_sim_X = monte_carlo_arima_pro(
    best_model_X, X_data_process_df["Price"].iloc[-1], steps=826, simulations=1000
)

monte_carlo_sim_Y = monte_carlo_arima_pro(
    model_Y_updated, last_aligned_price_Y, steps=826, simulations=1000
)

monte_carlo_sim_Z = monte_carlo_arima_pro(
    model_Z_updated, last_aligned_price_Z, steps=826, simulations=1000
)
```

```python
monte_carlo_team_1 = 0.2*monte_carlo_sim_X + 0.8*monte_carlo_sim_Y
monte_carlo_team_2 = (monte_carlo_sim_X + monte_carlo_sim_Y + monte_carlo_sim_Z) / 3
print("Costo preliminar del equipo 1:", np.mean(monte_carlo_team_1))
print("Costo preliminar del equipo 2:", np.mean(monte_carlo_team_2))
```

```python
def adf_test(series, name):
    result = adfuller(series.dropna())
    print(f"{name} - p-value:", result[1])
def adf_test_ct(series, name):
    result = adfuller(series, regression="ct")
    print(f"{name} - p-value:", result[1])

adf_test(X_data_process_df["Price"], "X Price")
adf_test(Y_data_process_df["Price"], "Y Price")
adf_test_ct(Y_data_process_df["Price"], "Y Price ct")
adf_test(Z_data_process_df["Price"], "Z Price")
```

Con esto encontramos que las materias primas X y Z no son estacionarias y la materia prima Y al parecer es estacionaria, pero si luego la tratamos con tendencia no se sostiene la hipótesis al 5%. Lo que nos lleva a considerar las materias primas como L(1), diferenciación de primer orden.

En este caso usar Arima con Drift es ideal frente a ETS o SARIMA, principalmente gracias a la evaluación del p-value y al enfoque que requiere este caso

#### Modelos preliminares

```python
def select_arima_levels(series, p_values=[0,1,2], q_values=[0,1,2]):
    results = []

    for p in p_values:
        for q in q_values:
            try:
                model = ARIMA(series, order=(p,1,q), trend="t")
                fitted = model.fit(method="innovations_mle")
                results.append((p,q,fitted.aic))
            except:
                continue
    data_df = pd.DataFrame(results, columns=["p","q","AIC"])
    return data_df.sort_values("AIC").reset_index(drop=True)
```

```python
X_aic_levels = select_arima_levels(X_data_process_df["Price"])
Y_aic_levels = select_arima_levels(Y_data_process_df["Price"])
Z_aic_levels = select_arima_levels(Z_data_process_df["Price"])

print(X_aic_levels.head())
print(Y_aic_levels.head())
print(Z_aic_levels.head())
```

Con estos resultados, concluimos que:
1. Materia prima X: Usaremos p:2 y q:2.
2. Materia prima Y: Usaremos p:1 y q:1
3. Materia prima Z: Usaremos p:2 y q:2

Debido que no hay una diferencia no significativa que nos ponga a dudar a tomar varios p y q.

```python
arima_X = ARIMA(
    X_data_process_df["Price"],
    order=(2,1,2),
    trend="t"
).fit(method="innovations_mle")

arima_Y = ARIMA(
    Y_data_process_df["Price"],
    order=(1,1,1),
    trend="t"
).fit(method="innovations_mle")

arima_Z = ARIMA(
    Z_data_process_df["Price"],
    order=(2,1,2),
    trend="t"
).fit(method="innovations_mle")
```

#### Ejecución para pronosticar 1 año de las materias primas

```python
def evaluate_arima_levels(series, order, test_days=252):
    train = series[:-test_days]
    test = series[-test_days:]
    model = ARIMA(train, order=order, trend="t").fit(method="innovations_mle")
    forecast = model.forecast(steps=test_days)

    mae = mean_absolute_error(test, forecast)
    rmse = np.sqrt(mean_squared_error(test, forecast))
    mape = np.mean(np.abs((test - forecast) / test)) * 100

    return mae, rmse, mape
```

```python
metrics_X = evaluate_arima_levels(
    X_data_process_df["Price"],
    order=(2,1,2)
)

metrics_Y = evaluate_arima_levels(
    Y_data_process_df["Price"],
    order=(1,1,1)
)

metrics_Z = evaluate_arima_levels(
    Z_data_process_df["Price"],
    order=(2,1,2)
)

print("Materia prima X - MAE, RMSE, MAPE:", metrics_X)
print("Materia prima Y - MAE, RMSE, MAPE:", metrics_Y)
print("Materia prima Z - MAE, RMSE, MAPE:", metrics_Z)
```

```python
def plot_evaluation(series, order, test_days=252, title=""):
    train = series[:-test_days]
    test = series[-test_days:]

    model = ARIMA(train, order=order, trend="t").fit(method="innovations_mle")
    forecast = model.forecast(steps=test_days)
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=series.index,
        y=series,
        name="Histórico"
    ))
    fig.add_trace(go.Scatter(
        x=test.index,
        y=forecast,
        name="Evaluación pronóstico"
    ))
    fig.update_layout(title=title, template="plotly_white")
    fig.show()
```

```python
plot_evaluation(X_data_process_df["Price"], (2,1,2), title="Materia prima X")
plot_evaluation(Y_data_process_df["Price"], (1,1,1), title="Materia prima Y")
plot_evaluation(Z_data_process_df["Price"], (2,1,2), title="Materia prima Z")
```
